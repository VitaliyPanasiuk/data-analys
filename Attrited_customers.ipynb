#Удаление ненужных столбцов и приведение названий к нижнему регистру без пробелов




```python
df.drop('CLIENTNUM',axis=1,inplace=True)
```

```python
df.drop('Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2',axis=1,inplace=True)
```

```python
df.drop('Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.columns = df.columns.str.lower()
```

```python
df.head()
```

Замена Attrited Customer и Existing Customer на числв 1 и 0


```python
(df.groupby(['attrition_flag']).agg(['count']))
```

```python
df['attrition_flag'] = df['attrition_flag'].replace('Attrited Customer',1).replace('Existing Customer',0)
```

```python
df.head()
```

#Анализ

```python
df['customer_age_group'] = pd.cut(df['customer_age'],5)
```

```python
df.head()
```

```python
df.groupby('customer_age_group')['attrition_flag'].agg(['count','mean'])
```

```python
df['customer_age_group2'] = pd.qcut(df['customer_age'],5)
```

```python
df.groupby('customer_age_group2')['attrition_flag'].agg(['count','mean'])
```

```python
df.groupby('customer_age_group2')['attrition_flag'].mean().plot(rot=45,ylim=0,grid=True)
```

с увеличением возрасата процент людей отказавшихся от карт увеличивается, но возмножно влияние другой переменной так как рост нелинейный


```python
df.groupby(['education_level','customer_age_group2'])['attrition_flag'].agg(['count','mean'])
```

Влияние возраста остаеться, чем старше клиенты тем чаще они октазыватся от карт за исключением возраста 44-53

```python
df.head()
```

```python
df.groupby('gender')['attrition_flag'].agg(['count','mean'])
```

выборка мужчин и женщин приблизителбно равная но видно что женщины чаще отказыабтся от карт

```python
df.groupby(['gender','customer_age_group2'])['attrition_flag'].agg(['count','mean'])
```

```python
desired_order = ['Unknown','Less than $40K','$40K - $60K', '$60K - $80K', '$80K - $120K',  '$120K +']
df.groupby('income_category')['attrition_flag'].mean().reindex(desired_order).plot(rot=45,ylim=0,grid=True)
```
